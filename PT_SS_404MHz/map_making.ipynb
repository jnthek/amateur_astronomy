```python
import numpy as np
import healpy as hp

dec_list = np.loadtxt("map_data.txt", usecols=0, delimiter=" ")
ra_list = np.loadtxt("map_data.txt", usecols=1, delimiter=" ")
T_K_list = np.loadtxt("map_data.txt", usecols=2, delimiter=" ")
```

```python
nside = 8
npix = hp.nside2npix(nside)
map_array = np.zeros(npix)

for i in range(len(dec_list)):
    pix_req = hp.ang2pix(nside, ra_list[i], dec_list[i], lonlat=True)
    map_array[pix_req] = T_K_list[i]
```

```python
hp.mollview(map_array, coord='EG')
```
